```python
ipl['date'] = pd.to_datetime(ipl['date'])

ipl['season_year'] = ipl['date'].dt.year
```

```python
season_batter_runs = (
    ipl.groupby(['season_year', 'batter'])['runs_batter']
       .sum()
       .reset_index()
)
```

```python
orange_cap = season_batter_runs.loc[
    season_batter_runs.groupby('season_year')['runs_batter'].idxmax()
]

orange_cap
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15,6))

sns.barplot(
    data=orange_cap,
    x='season_year',
    y='runs_batter'
)

plt.title("Orange Cap Winners by Season")
plt.xlabel("Season")
plt.ylabel("Runs")

plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(16,7))

bars = plt.bar(
    orange_cap['season_year'],
    orange_cap['runs_batter']
)

for bar, player in zip(bars, orange_cap['batter']):
    plt.text(
        bar.get_x() + bar.get_width()/2,
        bar.get_height() + 5,
        player,
        rotation=90,
        ha='center',
        fontsize=8
    )

plt.title("Orange Cap Winners Season Wise")
plt.xlabel("Season")
plt.ylabel("Runs")

plt.show()
```

```python
plt.savefig(
    "images/orange_cap_winners.png",
    dpi=300,
    bbox_inches="tight"
)
```

```python
bat_stats = (
    ipl.groupby('batter')
    .agg({
        'runs_batter':'sum',
        'balls_faced':'sum'
    })
)
```

```python
bat_stats['SR'] = (
    bat_stats['runs_batter']
    /
    bat_stats['balls_faced']
) * 100
```

```python
bat_stats = bat_stats[
    bat_stats['balls_faced'] > 500
]
```

```python
top_sr = (
    bat_stats
    .sort_values('SR', ascending=False)
    .head(10)
)

top_sr['SR'].plot(
    kind='barh',
    figsize=(10,6)
)

plt.title("Best Strike Rates")

plt.show()
```

```python
plt.savefig(
    "images/best_strike_rates.png",
    dpi=300,
    bbox_inches="tight"
)
```
